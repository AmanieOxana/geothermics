# A deep geothermal doublet

Here we evaluate the sample problem from the lecture (part 5) which is related to the previous notebook but uses simplified assumption. Hence, it is possible to solve it (more or less) on paper completely, nevertheless you will find the solution. 


```python
# import some libraries  
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Parameters

rho= 998
#kg/m^3

cp = 4179
#J/kg/K

flow_rate = 0.1
#m^3/s

ty=np.linspace(1,30)
dT=(100-8/3*ty)

# Plot Temperature difference

plt.plot(ty,dT,'-', color='red')


plt.xlabel('Time (years)')
plt.ylabel('Temperature (°C)')

plt.grid(True)
```

```python
#Thermal power produced from wells
pth = cp * rho * flow_rate * dT


plt.plot(ty,pth/1e6,'-', color='red')


plt.xlabel('Time (years)')
plt.ylabel('Thermal power (MW)')
plt.grid(True)
```

```python
#Efficiency
eta = 0.1;
# Installation consumption (MW)
ic = 0.5e6;
# Electrical power
pel = eta * pth - ic
plt.plot(ty, pel/1e6,'-', color='red')

plt.xlabel('Time (years)')
plt.ylabel('Electrical power (MW)')
plt.grid(True)
```

```python
# Calculating the energy

y2h=8760;

Eel=eta*cp * rho * flow_rate*(100*ty-4/3*ty**2)/1000-ic*ty/1000 # in kWa
# Conversion in kWh
Eel=Eel*y2h


plt.plot(ty, Eel,'-', color='red')


plt.xlabel('Time (years)')
plt.ylabel('Electrical energy (kWh)')
plt.grid(True)

```

```python
#Payment in Germany - EEG (Euro/kWh)
EEG=0.25

#Investment (Mio. Euro) (for plotting as an array with the same length as pel1)
invest = np.ones(len(pel))*46;

# Gain in Mio EUR
G1=Eel*EEG/1e6


# plotting the gain (in Mio EUR)
plt.plot(ty,G1,'-', color='red')
plt.plot(ty,invest,'--', color='black',label='Investment')


plt.legend()

plt.xlabel('Time (years)')
plt.ylabel('Gain (Mio. Euro))')

plt.grid(True)
```
